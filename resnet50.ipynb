```python
import torch
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms
from PIL import Image
import albumentations as A

# mean = [0.0, 0.0, 0.0]
# std = [1.0, 1.0, 1.0]
# max_pixel_value = 255.0

# Tạo lớp con của Dataset để tải dữ liệu huấn luyện
class MyDataset(Dataset):
    def __init__(self, data, labels, transform=None):
        self.data = data
        self.labels = labels
        self.transform = transform

    def __getitem__(self, index):
        x = self.data[index]
        x = Image.open(x).convert('RGB')
#         x = cv2.resize(x, (256, 256))
#         x = torch.stack((x,)*3, axis=-3)
    
        y = self.labels[index]
        

        if self.transform:
            x = self.transform(x)

        return x, y

    def __len__(self):
        return len(self.data)

# Định nghĩa các phép biến đổi cho dữ liệu ảnh
transform = transforms.Compose([
    transforms.Resize((256, 256)),  # Resize ảnh về kích thước 256x256
#     transforms.CenterCrop(224),
#     transforms.RandomHorizontalFlip(),  # Lật ảnh theo chiều ngang ngẫu nhiên
#     transforms.RandomRotation(15),  # Xoay ảnh ngẫu nhiên trong khoảng -10 đến 10 độ
    transforms.ToTensor(),  # Chuyển ảnh sang định dạng Tensor
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))  # Chuẩn hóa giá trị pixel về khoảng -1 đến 1
])


# Tạo đối tượng Dataset và DataLoader cho dữ liệu huấn luyện
train_dataset = MyDataset(X_train, y_train, transform=transform)
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)  # Batch size và shuffle tùy chọn

test_dataset = MyDataset(X_test, y_test, transform=transform)
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=True)  # Batch size và shuffle tùy chọn
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision.models import resnet50

class ResNet50(nn.Module):
    def __init__(self, num_classes=2):
        super(ResNet50, self).__init__()
        self.resnet = resnet50(pretrained=True)
        self.fc1 = nn.Linear(1000, 512)
        self.fc2 = nn.Linear(512, 256)
        self.fc3 = nn.Linear(256, num_classes)

    def forward(self, x):
        x = self.resnet(x)
        x = x.view(x.size(0), -1)
        x = nn.functional.relu(self.fc1(x))
        x = nn.functional.relu(self.fc2(x))
        x = self.fc3(x)
        return x
```

```python
import torch
import torch.nn as nn
from torchvision.models import resnet50
import torchvision.models as models
import timm

 
model = ResNet50(num_classes=2)
model = model.to('cuda')
num_epochs = 6

# Cấu hình hàm mất mát
criterion = nn.CrossEntropyLoss()

# Cấu hình bộ tối ưu hóa
optimizer = torch.optim.SGD(model.parameters(), lr=0.001, momentum=0.9)
# optimizer = optim.Adam(model.parameters(), lr=0.01)

# Vòng lặp huấn luyện
for epoch in range(num_epochs):
    model.train()
    train_loss = 0.0
    train_acc = 0.0
    total_samples = 0
    for inputs, labels in train_loader:
        # Chuyển đổi dữ liệu vào GPU
        inputs = inputs.to('cuda')
        labels = labels.to('cuda')

        # Xóa gradients
        optimizer.zero_grad()

        # Tính đầu ra từ mô hình
        outputs = model(inputs)

        # Tính hàm mất mát
        loss = criterion(outputs, labels)

        # Lan truyền ngược và cập nhật gradients
        loss.backward()
        optimizer.step()

        # Cập nhật tổng loss và số lượng mẫu
        train_loss += loss.item() * inputs.size(0)
        total_samples += inputs.size(0)

        # Tính accuracy
        _, predicted = outputs.max(1)
        train_acc += predicted.eq(labels).sum().item()

    # Tính giá trị trung bình của loss và accuracy trong epoch hiện tại
    train_loss /= total_samples
    train_acc /= total_samples

    # In thông tin quá trình huấn luyện
    print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {train_loss:.4f}, Acc: {train_acc:.4f}')
    
    model.eval()
    # Đánh giá trên tập test
    test_loss = 0.0
    test_acc = 0.0
    total_samples = 0

    for inputs, labels in test_loader:
        # Chuyển đổi dữ liệu vào GPU
        inputs = inputs.to('cuda')
        labels = labels.to('cuda')

        # Tính đầu ra từ mô hình
        outputs = model(inputs)

        # Tính hàm mất mát
        loss = criterion(outputs, labels)

        # Cập nhật tổng loss và số lượng mẫu
        test_loss += loss.item() * inputs.size(0)
        total_samples += inputs.size(0)

        # Tính accuracy
        _, predicted = outputs.max(1)
        test_acc += predicted.eq(labels).sum().item()

    # Tính giá trị trung bình của loss và accuracy trên tập test
    test_loss /= total_samples
    test_acc /= total_samples

    # In kết quả đánh giá trên tập test
    print(f'Test Loss: {test_loss:.4f}, Test Acc: {test_acc:.4f}')
```
